```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_correlation_map( db ):

    corr = db.corr()

    s , ax = plt.subplots( figsize =( 20 , 20 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )

db
```

```python
plot_correlation_map(db)
```

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import metrics
plt.scatter(db.price,db.sqft_living,color="r")
plt.title("Linear Regression")
plt.ylabel("price")
plt.xlabel("sqft_living")
plt.show()
```

```python
#question2

'''A correlation matrix is a table showing correlation coefficients between variables. Each cell in the table shows 
the correlation between two variables
This correlation matrix indicates that:
-price is very highly correlated with bathrooms, sqft_living,grade, sqft_above and sqft_living15 (value babove 0.5)
and highly correlated with bedrooms, floors,waterfront,view, sqft_basement and lat(value between 0.25 and 0.5 ) 
'''
```

```python
#question 3
#extract x and y from our data
x=db["sqft_living"].values[:,np.newaxis]
y=db["price"].values

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.35,random_state=40) #splitting data with test size of 35%
#question4
model=LinearRegression()   #build linear regression model
model.fit(x_train,y_train)  #fitting the training data
predicted=model.predict(x_test) #testing our model’s performance


```

```python
#question5
plt.scatter(x,y,color="r")
plt.title("Linear Regression")
plt.ylabel("sqft_living")
plt.xlabel("price")
plt.plot(x,model.predict(x),color="k")
plt.show()
```

```python
#question5
print("MSE", mean_squared_error(y_test,predicted))
print("R squared", metrics.r2_score(y_test,predicted))
```
