```python
def plot_correlation_map( db ):

    corr = db.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
```

```python
plot_correlation_map(db)
```

```python
'''A correlation matrix is a table showing correlation coefficients between variables. Each cell in the table shows 
the correlation between two variables
This correlation matrix indicates that:
-SibSp and Parch are weakly positively correlated (0,41 which is the most high positively correlated value)
-Pclass and Fare are strongly negatively correlated(-0.55 which is the most high negatively correlated value)
-Pclass and parch are basically not correlated.(0.018: There is very little association between Pclass and Parch). The same thing for 
the cell in the matrix with value -0.1 and 0.1
'''
```
